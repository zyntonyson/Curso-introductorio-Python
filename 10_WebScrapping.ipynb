```

website_url = requests.get('https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area').text
soup = BeautifulSoup(website_url,'lxml')
```

```
print(soup.prettify())  # Preview WP locate tags
```

```
# Select table
My_table = soup.find('table',{'class':'wikitable sortable'})  # Vistazo al HTML
```

```
# Countries
Countries=[]
Population=[]
rows = My_table.find_all('tr')
for row in rows:
  if row.has_attr('class'):
    continue
  cols=row.find_all('td')
  for i,col in enumerate(cols):
    if i==1:
      aux=col.find('a')
      if aux!=None:
        Countries.append(aux['title'])
    if i==2:
      Population.append(int(col.text.replace('\n','').replace(',','')))
```

```
tabla_df=pd.DataFrame(zip(Countries,Population), columns=['Country','Area.'])
```

Scrapp on : https://www.imdb.com/list/ls026392657/


```
mcu_list=requests.get('https://www.imdb.com/list/ls026392657/').text
marvelsoup=BeautifulSoup(mcu_list,'lxml')
```

```
marvelsoup
```

```
list_mcu=marvelsoup.find('div',{'class':'lister list detail sub-list'})
```

```
list_mcu
```

```
# Obtener año de lanzamiento
year_release=list_mcu.find_all('span',{'class':'lister-item-year text-muted unbold'})
```

```
release_Y=[]
for  year in year_release:
  release_Y.append(int(year.get_text().replace('(','').replace(')','')))
```

```
# Obtener titulo de la pelicula
movies=list_mcu.find_all('h3',{'class':'lister-item-header'})
```

```
title_movie=[]
#Lista  de películas
for movie in movies:
  #print(movie.find('a').text)
  title_movie.append(movie.find('a').text)
```

```
## Obtener reparto de la película

# Primero obtenemos la url de cada pelicula para acceder a su reparto
url_movie=[]
for movie in movies:
  url_movie.append(movie.find('a')['href'])
 
```

```
#Función que extrae el reparto de una pelicula

def get_cast_movie(url_movie):
  url='https://www.imdb.com'+url_movie
  cast_list=['dummy','dummy']
  movie_req=requests.get(url).text
  movie=BeautifulSoup(movie_req,'lxml')
  table_cast=movie.find('table',{'class':'cast_list'})
  nActor=0
  if table_cast!=None:
    actors=table_cast.find_all('tr')
    for actor in actors:
      if actor.has_attr('class') & nActor<15:
        name_actor=actor.find('td',{'class':'primary_photo'})
        if name_actor != None:
          name_actor=name_actor.find('a')
          if name_actor !=None:
            name_actor=name_actor.find('img')
            if name_actor != None:
              cast_list.append(name_actor.get('title'))
              nActor+=1
  return ",".join(cast_list)  
```

```
#Elenco
cast_mcu_movie=[get_cast_movie(movie) for movie in url_movie]
```

```
# Obtener reseña
movies_sipnosis=list_mcu.find_all('p',{'class':''})
```

```
sip_mov=[]
for sip in movies_sipnosis:
  #print(sip.text)
  sip_mov.append(sip.text.replace('\n',''))
```

```
# Obtener puntuacion
movies_rate=list_mcu.find_all('div',{'ipl-rating-star small'})
```

```
rates=[]
for rate in movies_rate:
  rates.append(rate.get_text().replace('\n',''))
```

```
mcu_df=pd.DataFrame(zip(title_movie,sip_mov,rates,cast_mcu_movie,release_Y), columns=['Title','Sipnosis','RateIMDB','Cast','Release'])
```

Ahora vamos a obtener las peliculas donde han participado actores de Marvel y obtener el reparto de los mismo

```
# Obtener el elenco de peliculas de marvel
cast_list_total=[]
url_cast=[]

def cast_movie_url(url_movie,cast_list=cast_list_total,url_cast=url_cast):
  url='https://www.imdb.com'+url_movie
  movie_req=requests.get(url).text
  movie=BeautifulSoup(movie_req,'lxml')
  table_cast=movie.find('table',{'class':'cast_list'})
  actors=table_cast.find_all('tr')
  for actor in actors:
    if actor.has_attr('class'):
      actor_name=actor.find('td',{'class':'primary_photo'}).find('a').find('img').get('title')
      if actor_name not in cast_list_total:
        cast_list_total.append(actor_name)
        url_cast.append(actor.find('td',{'class':'primary_photo'}).find('a')['href'])  
```

```
#Obtener elenco con link a pagina personal

for url in url_movie:
  cast_movie_url(url)
```

```
# Obtenemos el listado de peliculas donde haya participado algún miembro del elenco de Marvel

all_movies_batch=[]
all_movies_batch_url=[]

def moviesForActor(urlActor,all_movies_batch=all_movies_batch,all_movies_batch_url=all_movies_batch_url):
  actor_req=requests.get('https://www.imdb.com'+urlActor).text
  actor_page=BeautifulSoup(actor_req,'lxml')
  table_filmografy=actor_page.find('div',{'class':'filmo-category-section'})
  raw_filmo=table_filmografy.find_all('div',{'class':'filmo-row odd'}) +table_filmografy.find_all('div',{'class':'filmo-row even'})
  for film in raw_filmo:
    movie_name=film.find('b').find('a').get_text()
    if movie_name not in all_movies_batch:
      all_movies_batch.append(movie_name)
      all_movies_batch_url.append(film.find('b').find('a')['href'])

```

```
for url in url_cast:
  moviesForActor(url)
```

```
nmovies=0
for movie in all_movies_batch:
  print(movie, end= ' ')
  nmovies+=1
  if nmovies%5==0:
    print()
```
